Makes a contour plot showing some statistic and use stippling to show statistically significant regions.

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
```

Make some dummy data

```python
# invent some numbers, turning the x and y arrays into simple
# 2d arrays, which make combining them together easier.
x = np.linspace(-3, 5, 150).reshape(1, -1)
y = np.linspace(-3, 5, 120).reshape(-1, 1)
z = np.cos(x) + np.sin(y)

p = np.cos(x**2) + np.sin(y)
p = p - p.min() 
p = p / (p.max() - p.min())

# we no longer need x and y to be 2 dimensional, so flatten them.
x, y = x.flatten(), y.flatten()
```

Plo

```python
fig = plt.figure(figsize=(7,7))

map_proj = ccrs.NorthPolarStereo()

ax = fig.add_subplot(projection=map_proj)
ax.set_extent([-180., 180., 60., 90.], ccrs.PlateCarree())
ax.coastlines()

ax.contourf(x*1e6, y*1e6, z, transform=map_proj)
ax.contourf(x*1e6, y*1e6, p, colors='none', hatches=['..'], levels=[.9, 1.])
```
