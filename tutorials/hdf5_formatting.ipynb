```python
import h5py
import xarray as xr
import numpy as np
```

```python
root = h5py.File('test.h5', 'w')
```

Create some groups

```python
grp1 = root.create_group("group_1")
grp2 = root.create_group("group_2")
```

Create some Datasets in those groups

```python
nx1 = 4
ny1 = 3
grp1["data1"] = np.ones((nx1, ny1), 'f')
grp1["data2"] = np.ones((nx1, ny1), 'f')

nx2 = 5
ny2 = 3
grp2["data3"] = np.ones((nx2, ny2), 'i')
```

Create some dimension variables.  The dimension scales need to be in the group so that they can found by `xarray`. --- 

```python
grp1['x'] = np.arange(nx1)
grp1['x'].make_scale('x')
grp1['y'] = np.arange(ny1)
grp1['y'].make_scale('y')
```

Attach scales to the group_1 data

```python
grp1["data1"].dims[0].attach_scale(grp1['x'])
grp1["data1"].dims[1].attach_scale(grp1['y'])

grp1["data2"].dims[0].attach_scale(grp1['x'])
grp1["data2"].dims[1].attach_scale(grp1['y'])
```

```python
list_of_objects = []
root.visit(list_of_objects.append)
list_of_objects
```

```python
root.close()
```

Open group_1 with `xarray`
